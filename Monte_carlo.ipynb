## True parameters

```python
beta_true = np.array([3.0, -0.05, 0.5], dtype=float)
lambda_true = np.array([0.2, 0.3], dtype=float)

theta0 = np.array([1.0, -0.35, 0.35, 1.0, 0.15], dtype=float)
```

## Monte Carlo

##### Setup

```python
R = 250
T = 50
N = 10000
J = 7
```

##### Choice + MSW

```python
selected_moments = ["choice", "two_search", "one_search"]

mc_res_2 = est.monte_carlo_gmm_one_step(
    R=R,
    T=T,
    N=N,
    J=J,
    beta_true=beta_true,
    lambda_true=lambda_true,
    theta0=theta0,
    selected_moments=selected_moments,
    seed_start=1000
)

df_2 = pd.DataFrame({
    "replication": [r["replication"] for r in mc_res_2],
    "success": [r["success1"] for r in mc_res_2],
    "Objective": [r["fun1"] for r in mc_res_2],
    "beta_const": [r["theta1"][0] for r in mc_res_2],
    "beta_price": [r["theta1"][1] for r in mc_res_2],
    "beta_star": [r["theta1"][2] for r in mc_res_2],
    "lambda_const": [r["theta1"][3] for r in mc_res_2],
    "lambda_pos": [r["theta1"][4] for r in mc_res_2]
})
```

##### Choice + MSW + Searched product

```python
selected_moments = ["choice", "two_search", "one_search", "searched_product"]

mc_res_3 = est.monte_carlo_gmm_one_step(
    R=R,
    T=T,
    N=N,
    J=J,
    beta_true=beta_true,
    lambda_true=lambda_true,
    theta0=theta0,
    selected_moments=selected_moments,
    seed_start=1000
)

df_3 = pd.DataFrame({
    "replication": [r["replication"] for r in mc_res_3],
    "success": [r["success1"] for r in mc_res_3],
    "Objective": [r["fun1"] for r in mc_res_3],
    "beta_const": [r["theta1"][0] for r in mc_res_3],
    "beta_price": [r["theta1"][1] for r in mc_res_3],
    "beta_star": [r["theta1"][2] for r in mc_res_3],
    "lambda_const": [r["theta1"][3] for r in mc_res_3],
    "lambda_pos": [r["theta1"][4] for r in mc_res_3]
})
```

##### Choice + MSW + Expected number of searches

```python
selected_moments = ["choice", "two_search", "one_search", "exp_search"]

mc_res_4 = est.monte_carlo_gmm_one_step(
    R=R,
    T=T,
    N=N,
    J=J,
    beta_true=beta_true,
    lambda_true=lambda_true,
    theta0=theta0,
    selected_moments=selected_moments,
    seed_start=1000
)

df_4 = pd.DataFrame({
    "replication": [r["replication"] for r in mc_res_4],
    "success": [r["success1"] for r in mc_res_4],
    "Objective": [r["fun1"] for r in mc_res_4],
    "beta_const": [r["theta1"][0] for r in mc_res_4],
    "beta_price": [r["theta1"][1] for r in mc_res_4],
    "beta_star": [r["theta1"][2] for r in mc_res_4],
    "lambda_const": [r["theta1"][3] for r in mc_res_4],
    "lambda_pos": [r["theta1"][4] for r in mc_res_4]
})
```

## Visualization

```python
plot_dfs = {
    "C+OS+TS": df_2,
    "C+OS+TS+SP": df_3,
    "C+OS+TS+ES": df_4,
}


parameter_name = "beta_const"
true_value = beta_true[0]

all_estimates = np.concatenate([
    df[parameter_name].dropna().to_numpy()
    for df in plot_dfs.values()
])

x_grid = np.linspace(-2, 6, 300)

colors = ['#f5ad76', '#e76f51', '#a62626']
plt.figure(figsize=(7, 5))

for (label, df), color in zip(plot_dfs.items(), colors):
    estimates = df[parameter_name].dropna().to_numpy()
    kde = gaussian_kde(estimates)
    plt.plot(x_grid, kde(x_grid), label=label, color=color)

plt.axvline(true_value, linestyle="--", color="black", label="True value")

plt.xlabel("Estimate")
plt.ylabel("Density")
plt.title(f"Distribution of estimates: {parameter_name}")
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```

```python
for name, df in plot_dfs.items():
    estimates = df[parameter_name].dropna().to_numpy()
    np.savetxt(f"estimates_{name}_{parameter_name}_low.txt", estimates)
    
    mean = np.mean(estimates)
    bias = np.mean(estimates) - true_value
    sd = np.std(estimates, ddof=1)
    length = len(estimates)
    mcse = sd / np.sqrt(length)
    
    print(f"Parameter name: {parameter_name}")
    print(f"Moment Set: {name}")
    print(f"True: {true_value}")
    print(f"R: {length}")
    print(f"Mean: {mean:.4f}, Bias: {bias:.4f}, MCSE: {mcse:.4f}, SD: {sd:.4f}\n")
```
